```python
import pandas as pd
import matplotlib.pyplot as plt
file = 'NSDUH_2019.DTA'
df = pd.read_stata(file)
df.head()
```

```python
df.loc[[10]]
```

```python
print(list(df))
print(len(df))
print(len(list(df)))
```

```python
col = 'yowrwrth'
col_name = '\"When problems worsen, feel totally worthless daily\" [Teenagers]'
freq = df[col].value_counts().rename_axis(col_name).reset_index(name='freq')
freq = freq.sort_values(by = col_name)[:2]
print(freq)
```

```python
ax = freq.plot.bar(x=col_name, y='freq', color="navy", rot=1)
```
