```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager
file = 'NSDUH_2019.DTA'
df = pd.read_stata(file)
df.head()
```

```python
#SET STYLE FOR REST OF NOTEBOOK
matplotlib.style.use('seaborn') #set stylesheet
matplotlib.rcParams['font.size'] = 20
matplotlib.rcParams['figure.figsize'] = (15, 12) #standardise figure size
matplotlib.rcParams['font.family'] = 'Verdana'
#matplotlib.rcParams['font.weight'] =  'bold' #<-consider adding this line (uncommented) if using a thin font
```

```python
#GET Fonts and write to file fonts.txt in same folder (NOTE: not every font will work)
fonts = font_manager.findSystemFonts(fontpaths = None, fontext = 'ttf')
file = open('fonts.txt', 'w')
file.write(str(fonts))
```

```python
col = 'yowrwrth'
col_name = '\"Surveyed Response\" [Teenagers]'
freq = df[col].value_counts().rename_axis(col_name).reset_index(name='freq')
freq = freq.sort_values(by = col_name)[:2]
print(freq)
```

```python
ax = freq.plot.bar(x = col_name, y='freq', color = ['navy', 'green'], rot=1, fontsize = 15)
plt.title('\"When problems worsen, feel totally worthless daily\" [Teenagers]', fontsize = 20)
plt.xlabel(col_name, fontsize=20)
plt.savefig('Congy_Bar4.png', dpi = 300)
```
